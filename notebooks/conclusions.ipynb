# Class Summary

## Part One: Social Media Platforms
In part one, we talked about some of the big questions surrounding social media ecosystems and the algorithms that manage information propogation among their users ([notebbok 2](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-2-social-media-tiktok.ipynb)), such as:

* Do social media algorithms actually influence the things that people beleive? Can they influence democratic processes?
* How do the design characteristics of a system impact information consumption?

We introduced a range of methods ([notebbok 2](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-2-social-media-tiktok.ipynb)) that researchers have used to examine the impacts of different kinds of algorithms on users, and to examine the dynamics of opaque algorithms. In particular, we focused on outside audits of content selection algorithm.

We began by diving into a recent study of TikTok's content selection algorithm:

* Baumann, F., Arora, N., Rahwan, I., & Czaplicka, A. (2025). [Dynamics of Algorithmic Content Amplification on TikTok](https://arxiv.org/pdf/2503.20231v1). arXiv preprint arXiv:2503.20231.

And an investigation of recent changes to [Reddit's terms of service for research practices](https://www.reddit.com/r/reddit/comments/1co0xnu/sharing_our_public_content_policy_and_a_new/) and the underlying motivations for these changes. 

In the context of such changes to data availability, we discussed Bluesky ([https://bsky.app/](https://bsky.app/)) -- a decentralized social media platform built around principles of openness. 

Bluesky offered an opportunity to [develop our skills](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-3-social-media-bluesky.ipynb) around interacting with platform APIs to retreive data. We used [Bluesky's HTTP API](https://docs.bsky.app/docs/category/http-reference) to explore the affordances of their data sharing model, and to practice retreiving data through http requests in Python, and using data to generate insights into user behavior and social structure on social media platforms ([notebbok 3](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-3-social-media-bluesky.ipynb)).

This section of the class introduced:
* **key topics and methods in social media platform research** 
* recent developments in the **modern landscape of data accessibility trends** (in the context of increasingly valuable large-scale AI model training datasets)
* **key concepts and skills** relevant to **retreiving data programmatically from remote servers** via API calls 

and offered opportunities to:
* **develop a broader understanding of the role that data analysis can play** in managing evolving information ecosystems and social technologies
* **refine practical data management skills** in Python 

## Part Two: Conversational AI Systems
In part two of the class, we began talking about conversational AI systems, their increasing prevelance in society, and analagous challenges that arise for analysts interested in understanding their impacts on information consumption and propogation, as well as the capabilities and limitations of such models themselves.

We began by introducing basic concepts and methods relevant to the evaluation of conversational AI systems, and the differences between static benchmarks and interactive evaluation ([notebook 4](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-4-llm-interactions.ipynb)).

We dived into a recent study introducing interactive evaluations of the pursuasiveness of frontier AI models:

* Salvi, F., Horta Ribeiro, M., Gallotti, R., & West, R. (2025). [On the conversational persuasiveness of GPT-4](https://www.nature.com/articles/s41562-025-02194-6). Nature Human Behaviour, 1-9.

We discussed the research paper in detail, and retreived the dataset assosciated with the paper to examine basic questions about the structure of the evaluations and the design of the study ([notebook 4](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-4-llm-interactions.ipynb)). Through this excersise, we introduced the [HuggingFace ecosystem for dataset storage and retreival](https://huggingface.co/docs/datasets/en/index), and had an opportunity to practice practical data management skills using [Seaborn](https://seaborn.pydata.org/) and [Pandas](https://pandas.pydata.org/) in Python.

In week five, we introduced key concepts around the logistics of hosting and accessing large language models, reviewed popular development platforms, and dived into the various frontier models available and their differences in pricing and capability ([notebook 5](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-5-llm-api-usage.ipynb)). 

We used this opportunity to develop our skills in interacting with remotely hosted models via API calls, using an app hosted on HuggingFace that routed requests to an instance of GPT2 hosted in the cloud. We perfomed basic evaluations of the model's capabilities through exploratory prompting, such as the model's capacity to understand simple concepts such as days of the week, or to perform cognitive tasks such as counting and evaluative judgements of short texts ([notebook 5](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-5-llm-api-usage.ipynb)).

# Hugging Face App

Our app is/was hosted at [https://huggingface.co/spaces/macss-css/llm-lab-two](https://huggingface.co/spaces/macss-css/llm-lab-two). In brief, our app is a Python-based web app hosted on [Huggingface spaces](https://huggingface.co/spaces/launch). Anyone can create a huggingface space using a free Huggingface account. The app provides endpoints for student post requests (e.g. `/generate`) and routes these requests to a model hosted locally in the space using the [transformers](https://huggingface.co/docs/transformers/en/index) library.   

 For [lab session five](https://github.com/ccs-ucb/CSS1AF25/blob/main/notebooks/notebook-5-llm-api-usage.ipynb), I used a free-tier CPU Basic container to serve a local instance of GPT2. For the Class Project, I upgraded the app to NVIDIA T4 Medium hardware to serve a local instance of Llama-2-7B, since this model requires a GPU and more memory. 

GPT-5 provides a nice overview of what a basic HF spaces app for this general purpose looks like:

`my-hf-space/`
* `main.py`
* `requirements.txt`
* `README.md`

Here's a MWE for `main.py`:

```python
from fastapi import FastAPI, Request
import requests
import os

app = FastAPI()

# You can set this in your Space’s "Secrets" tab as HF_TOKEN
HF_TOKEN = os.getenv("HF_TOKEN")
MODEL = "gpt2"

@app.get("/")
def root():
    return {"message": "Welcome to the GPT-2 relay Space!"}

@app.post("/generate")
async def generate(request: Request):
    data = await request.json()
    prompt = data.get("prompt", "")

    headers = {"Authorization": f"Bearer {HF_TOKEN}"}
    payload = {"inputs": prompt}

    response = requests.post(
        f"https://api-inference.huggingface.co/models/{MODEL}",
        headers=headers,
        json=payload
    )

    return response.json()
```
and the things you would need in `requirements.txt`:

```
fastapi
uvicorn
requests
```

There's much more to learn about how web apps work, but the template above may help as a conceptual starting point.

# Class Project Tips

* Consider being systematic about your identities. A well-designed experiment often manipulates a single variable whilst holding other variables constant. 
* Try different prompt structures. Can you get the model to give a definite answer to your question?
* Print the head of your responses dataset to a cell – so we can see what your dataset looks like
* Can you collect more than one response per identity from the model? 
* In the section “Analyzing and Visualizing the Impacts of User Identity”, try to include data in your responses.
